# Analysis

## Imports

```python
import pandas as pd
import scipy.stats as stats
```

Imports required libraries for this notebook

## Data

```python
mergedDataLocation = r"ModifiedData/ExportedData.csv"
newMergedOutput = r"ModifiedData/GraphingData.csv"
```

Creates an easily editable variable with the location of the cleaned and merged dataset

```python
MergedDF = pd.read_csv(mergedDataLocation, index_col=0)
MergedDF
```

Loads the merged data into a Pandas DataFrame object and displays it

## Analysis

### Generate List-wise Data for linear modelling

```python
EconomicData = list(MergedDF["GDP per capita, PPP (constant 2021 international $)"])
HappyData = list(MergedDF["Life evaluation (3-year average)"])

pd.DataFrame({"GDP per capita":EconomicData,"Reported Happiness Score":HappyData})
```

Creates two lists, one containing GDP per Capita and the other containing the Reported Happiness Score, from the Merged Dataframe. The lists are then temporarily put into a new dataframe to be displayed with headers, for comparison against the larger DataFrame.

### Model Creation

```python
# As shown at https://how-to-data.org/how-to-fit-a-linear-model-to-two-columns-of-data/
model = stats.linregress(EconomicData, HappyData)

f"Model equation: y = {model.slope}x + {model.intercept}; a standard error of {model.stderr}, p-value of {model.pvalue}, and R-squared of {model.rvalue**2}"
```

Performs a Linear Regression and Outputs the line generated

### Prediction Generation

```python
MergedDF["Predicted Happiness"] = MergedDF["GDP per capita, PPP (constant 2021 international $)"].apply(lambda x : (model.slope * x) + model.intercept)

MergedDF
```

Adds a Prediction column to the MergedDF for graphing

### Get Covariance and correlation of Per Capita GDP and Happiness Score

```python
gdpHappinessCovariance = MergedDF["GDP per capita, PPP (constant 2021 international $)"].cov(MergedDF["Life evaluation (3-year average)"])
gdpHappinessCorrelation = MergedDF["GDP per capita, PPP (constant 2021 international $)"].corr(MergedDF["Life evaluation (3-year average)"])
f"GDP per Capita and Reported Happiness Score have a covariance of {gdpHappinessCovariance} and a correlation of {gdpHappinessCorrelation}"
```

Gets correlation and covariance of GDP per Capita and Happiness Score, showing a strong positive correlation of 0.7345573178610736
